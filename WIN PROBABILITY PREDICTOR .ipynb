# NOW THIS IS A CLASSIFICATION PROBLEM BUT WE HAVE THE CHOOSE THAT TYPE OF MODEL WHICH CAN GIVE RESULT IN PROBABILITY .....
# AGAIN WE DONT NEED THESE VAST DATA ....
# WE HAVE TO EXTRACT OUR NECESSARY INFORMATION FROM THOSE TWO DATASETS ON WHICH OUR MODEL ACTUALLY RUNS

```python
# WE WANT TO DETERMINE THE TOTAL RUNS SCORED IN BOTH INNINGS
total_score_df=delivery.groupby(['ID','innings']).sum()['total_run'].reset_index()
total_score_df.head()
```

```python
#now we dont need the 2nd innings total score as we only want the first innings score which the 2nd team are chasing
total_score_df=total_score_df[total_score_df['innings']==1]
total_score_df.head()
```

```python
#now we merge this to our match dataset as in that dataset all the match details are there but the 1st innings total are not there
match_df=match.merge(total_score_df[['ID','total_run']])
# here we don't use any left join or right join condition as two data frame has a common coloumn name 'ID'
```

```python
match_df.head()
```

```python
# THERE ARE SUCH TEAMS WHICH DONT PLAY ANYMORE IN THE IPL 
# HERE WE NEED ONLY THOSE TEAMS WHICH CONTINUOUSLY PLAYING & ARE ACTIVE MODE RIGHT NOW
match_df['Team1'].unique()
```

```python
#list of teams which are currently active
teams=['Sunrisers Hyderabad','Mumbai Indians','Royal Challengers Bangalore','Kolkata Knight Riders','Punjab Kings','Chennai Super Kings','Rajasthan Royals','Delhi Capitals']
```

```python
# now we change the team names which change their names after some year of playing IPL
match_df['Team1']=match_df['Team1'].str.replace('Delhi Daredevils','Delhi Capitals')
match_df['Team2']=match_df['Team2'].str.replace('Delhi Daredevils','Delhi Capitals')

match_df['Team1']=match_df['Team1'].str.replace('Deccan Chargers','Sunrisers Hyderabad')
match_df['Team2']=match_df['Team2'].str.replace('Deccan Chargers','Sunrisers Hyderabad')

match_df['Team1']=match_df['Team1'].str.replace('Kings XI Punjab','Punjab Kings')
match_df['Team2']=match_df['Team2'].str.replace('Kings XI Punjab','Punjab Kings')

```

```python
# we want to take those rows where the teams are in our teams list
match_df=match_df[match_df['Team1'].isin(teams)]
match_df=match_df[match_df['Team2'].isin(teams)]
```

```python
match_df.shape
```

# WE CAN SEE THAT, AT FIRST WE HAVE ABOUT 950 MATCHES PRESENT IN THE MATCH DATASET.
# BUT NOW AS WE CLEAN THE DATASET WE GET ONLY 806 MATCHES

```python
# now we extract our necessary coloums from this match dataset & merge to the delivery dataset as the main features are present in the delivery dataset
match_df.head()
```

# NOW WE NEED THE FOLLOWING COLOUMNS ONLY:
# ID , CITY , WINNING TEAM , TOTAL RUN , TEAM 1

```python
match_df=match_df[['ID','City','WinningTeam','total_run','Team1']]
# WE MERGE THESE COLOUMNS WITH DELIVERY DATASET
delivery_df=match_df.merge(delivery,on='ID')
delivery_df.head()
```

```python
# now we are only concerned with ball by ball data of 2nd innings of all matches
# thats why we eclude all the rows where the innings is equals to 1
delivery_df=delivery_df[delivery_df['innings']==2]
delivery.shape
```

```python
# now we want the cuurent score per match wise & which updates in every ball
delivery_df['current_score']=delivery_df.groupby('ID').cumsum()['total_run_y']
delivery_df.head()
```

```python
# here we want to update the remaining runs after every ball
delivery_df['target']=delivery_df['total_run_x']+1
delivery_df['runs_left']=delivery_df['target']-delivery_df['current_score']
delivery_df.head()
```

```python
#now we want the number of remaining balls after every ball was pitched
delivery_df['remaining_balls']=120-(delivery_df['overs']*6+delivery_df['ballnumber'])
delivery_df.head()
```

```python
# now we want the number of remaining wickets 
wickets=delivery_df.groupby('ID').cumsum()['isWicketDelivery'].values
delivery_df['remaining_wickets']=10-wickets
delivery_df.tail()
```

```python
# NOW WE WANT TO CALCULATE CURRENT RUN RATE (CRR)
# CRR = RUNS/OVER
# crr = current score/[(120-remaining balls)/6]=current score * 6 / (120 - remaining balls)
delivery_df['crr']=delivery_df['current_score']*6/(120-delivery_df['remaining_balls'])
delivery_df.tail()
```

```python
# now we want to calculate required run rate
# required run rate = runs left / (remaining balls/6) = ( runs left * 6 ) / remaining balls
delivery_df['rrr']=(delivery_df['runs_left']*6)/delivery_df['remaining_balls']
delivery_df.head()
```

# now we want to create a coloumn named result which shows that if the 2nd batting team is winner or not
# if the chasing team wins then it takes the value 1 
# otherwise 0
# so we are doing this with the help of creating a function

```python
def result(row):
    return 1 if row['BattingTeam']==row['WinningTeam'] else 0
```

```python
delivery_df['result']= delivery_df.apply(result,axis=1)
delivery_df.head()
```

```python
final_df=delivery_df[['City','target','Team1','BattingTeam','runs_left','remaining_balls','remaining_wickets','crr','rrr','result']]

                     
            
```

```python
final_df.head()
```

# NOW WE HAVE ALL THE BALLS PLAYED IN ONE MATCH TOGETHER
# IT CAN CREATE A BIAS 
# THAT'S WHY WE SHUFFLE OUR FINAL DATADET

```python
# we shuffule the final dataset
final_df=final_df.sample(final_df.shape[0])
```

```python
final_df.dropna(inplace=True)
# We drop all the null Values
```

# We easily see that in the Required Run Rate Column there is still some Nan values , some Infinity values & some large values are present 
# We have to handle Them before Proceeding

# We can do one thing
# we drop those rows where remaining balls are equal to 0
# Because It creates Problem of taking Infinite Values

```python
# taking those rows only where remaining balls are not equal to 0
final_df=final_df[final_df['remaining_balls']!=0]
```

# NOW WE CAN CONCENTRATE ON MODEL BUILDING
# DATA CLEANING PART IS DONE


```python
# WE SPLIT TRAIN & TEST SET
x=final_df.drop(['result'],axis=1)
y=final_df['result']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=18)
```

```python
x_train
```

# now we can see that in our final dataset there are 3 coloumns which are not numerical in nature
# which can cause a problem to us
# that's why we have to apply ONE-HOT encoding method to convert those columns to numerical nature

```python
#one-hot encoding
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
trf=ColumnTransformer([('trf',OneHotEncoder(sparse=False,drop='first'),['BattingTeam','Team1','City'])],remainder='passthrough')
```

```python
# NOW WE APPLY OUR LOGISTIC REGRESSION MODEL
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
# WE USE PIPELINE BY THIS
```

```python
# We created the Pipeline
pipe=Pipeline(steps=[('step1',trf),('step2',LogisticRegression(solver='liblinear'))])
```

```python
# PREDICT TIME!!
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
# CHECKING ACCURACY OF THE MODEL
from sklearn.metrics import accuracy_score
a=accuracy_score(y_test,y_pred)
```

```python
print('The Acuuracy of Our Model:','%.2f'%(a))
```
